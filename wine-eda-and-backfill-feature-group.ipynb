### Cleaning

```python
wine_df = wine_df.dropna()
wine_df.info()
```

```python
wine_df['type'] = pd.get_dummies(wine_df['type'], dtype=int).drop('white',axis = 1)
wine_df.info()
```

```python
new_names = {}
for name in wine_df.columns:
    new_names[name] = name.lower().replace(' ', '_')
wine_df = wine_df.rename(columns=new_names)
wine_df.columns
```

### Insert our Wine DataFrame into a FeatureGroup

```python
wine_fg = fs.get_or_create_feature_group(
    name="wine",
    version=1,
    primary_key=list(wine_df.columns)[:-1],  # All columns exept 'quality'
    description="Wine quality dataset")
wine_fg.insert(wine_df)
```
